```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("YelpDataAnalysis") \
    .getOrCreate()

print(spark.version)
```

```python
# Loading the Yelp dataset (JSON format)
df = spark.read.json("yelp_academic_dataset_business.json")

# Showing the first 5 rows
df.show(5)
```

```python
import pandas as pd
# Convertng Spark DataFrame to Pandas for better reading
df.limit(10).toPandas()  
```

```python
#Checking for nested structure
df.printSchema()
```

```python
#Attributes 
df_attributes = df.select("business_id", "name", *[f"attributes.{c}" for c in df.schema["attributes"].dataType.names]).limit(10).toPandas()

df_attributes
```

```python
#Hours
df_hours = df.select("business_id", "name", *[f"hours.{c}" for c in df.schema["hours"].dataType.names]).limit(10).toPandas()

df_hours
```

```python

# Converting full dataset to Pandas for better readability
df_pandas = df.select(
    "business_id", "name", "city", "state", "address", "categories", "is_open", "stars", "review_count", 
    *[f"attributes.{c}" for c in df.schema["attributes"].dataType.names], 
    *[f"hours.{c}" for c in df.schema["hours"].dataType.names]  
).limit(10).toPandas()

# Displaying Pandas DataFrame
df_pandas
```

```python
#Counting Total Businesses
print(f"Total number of businesses: {df.count()}")
```

```python
#checking for dublicate business ids
print(f"Duplicate businesses: {df.count() - df.dropDuplicates(['business_id']).count()}")
```

```python
#Checking for Missing Values
from pyspark.sql.functions import col, sum

df.select([sum(col(c).isNull().cast("int")).alias(c) for c in df.columns]).show()
```

```python
# 2. Count of NULL values in each attribute (nested column handling)
attribute_columns = df.select("attributes.*").columns
null_counts_attributes = df.select([sum(col(f"attributes.{c}").isNull().cast("int")).alias(c) for c in attribute_columns])
null_counts_attributes_pd = pd.DataFrame(null_counts_attributes.collect(), columns=null_counts_attributes.columns).T.reset_index()
null_counts_attributes_pd.columns = ["Attribute", "Null Count"]
print(null_counts_attributes_pd.to_string(index=False))
```

```python
# 2b. Count of NULL values in `categories` column
df_categories = df.selectExpr("business_id", "split(categories, ', ') as category_list")
df_exploded = df_categories.selectExpr("business_id", "explode(category_list) as category")
null_counts_categories = df_exploded.groupBy("category").agg(sum(col("category").isNull().cast("int")).alias("null_count"))
null_counts_categories_pd = pd.DataFrame(null_counts_categories.collect(), columns=["Category", "Null Count"])
print(null_counts_categories_pd.to_string(index=False))
```

```python
# Select only 'business_id' and 'categories' columns where 'categories' is null
missing_categories = df.filter(col("categories").isNull()).select("business_id", "categories")

# Show the result
missing_categories.show()


```

```python
from pyspark.sql.functions import col, sum

# 3. Counting NULL values in `hours` column for each day of the week
hour_columns = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
null_counts_hours = df.select(
    *[sum(col(f"hours.{day}").isNull().cast("int")).alias(day) for day in hour_columns]
)

# Collecting and converting to Pandas DataFrame for better readability
null_counts_hours_pd = pd.DataFrame(null_counts_hours.collect(), columns=null_counts_hours.columns).T.reset_index()
null_counts_hours_pd.columns = ["Day", "Null Count"]

# Displaying the result
print(null_counts_hours_pd.to_string(index=False))
```

```python
# Check for stars outside the expected range (1.0 - 5.0)
outliers_stars = df_pandas[(df_pandas['stars'] < 1.0) | (df_pandas['stars'] > 5.0)]

# Display businesses with invalid stars ratings
print("Businesses with invalid stars ratings:")
print(outliers_stars[['business_id', 'name', 'stars']])
```

```python
# Checking review count outliers
q1 = df_pandas['review_count'].quantile(0.25)
q3 = df_pandas['review_count'].quantile(0.75)
iqr = q3 - q1

# Defining lower and upper bounds for review count
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

# Checking for businesses with review count outside the normal range
outliers_reviews = df_pandas[(df_pandas['review_count'] < lower_bound) | (df_pandas['review_count'] > upper_bound)]

# Display businesses with abnormal review counts
print("Businesses with abnormal review counts:")
print(outliers_reviews[['business_id', 'name', 'review_count']])
```

```python
# Businesses with high review count but low stars (rating <= 2.0 and review count > 100)
high_reviews_low_rating = df_pandas[(df_pandas['stars'] <= 2.0) & (df_pandas['review_count'] > 100)]

# Businesses with low review count but high stars (rating >= 4.5 and review count < 10)
low_reviews_high_rating = df_pandas[(df_pandas['stars'] >= 4.5) & (df_pandas['review_count'] < 10)]

# Display the businesses that might have inconsistencies
print("Businesses with high reviews but low stars or low reviews but high stars:")
print(pd.concat([high_reviews_low_rating[['business_id', 'name', 'stars', 'review_count']], low_reviews_high_rating[['business_id', 'name', 'stars', 'review_count']]]))
```

```python
from fuzzywuzzy import process

# Listing of standardized city names 
standardized_cities = ["Santa Barbara", "Affton", "Tucson", "Philadelphia", "Green Lane", "Ashland City", "Brentwood", "St. Petersburg", "Nashville"]

# Function to find the best match for each city
def normalize_city(city, choices, threshold=80):
    match = process.extractOne(city, choices)
    if match and match[1] >= threshold:
        return match[0]  # Return the matched city
    return city  # If no match is found or the match is below threshold, return the original

# Normalizing city names using fuzzy matching
df_pandas['normalized_city'] = df_pandas['city'].apply(lambda x: normalize_city(x, standardized_cities))

# Displaying the result
print(df_pandas[['city', 'normalized_city']].drop_duplicates())
```

```python
# Checking for outliers in latitude and longitude
outliers_lat = df[(df['latitude'] < -90) | (df['latitude'] > 90)]
outliers_lon = df[(df['longitude'] < -180) | (df['longitude'] > 180)]

print(f"Outliers in latitude: {outliers_lat}")
print(f"Outliers in longitude: {outliers_lon}")
```

```python
# Convert PySpark DataFrame to pandas DataFrame
df_pandas = df.toPandas()

# Z-score method
from scipy import stats
z_scores_lat = stats.zscore(df_pandas['latitude'].dropna())
z_scores_lon = stats.zscore(df_pandas['longitude'].dropna())

# Outliers based on Z-score (greater than 3 or less than -3)
outliers_lat_z = df_pandas[abs(z_scores_lat) > 3]
outliers_lon_z = df_pandas[abs(z_scores_lon) > 3]

print(f"Outliers in latitude (Z-score method): {outliers_lat_z}")
print(f"Outliers in longitude (Z-score method): {outliers_lon_z}")
```
