```python
from fuzzywuzzy import process

# Listing of standardized city names 
standardized_cities = ["Santa Barbara", "Affton", "Tucson", "Philadelphia", "Green Lane", "Ashland City", "Brentwood", "St. Petersburg", "Nashville"]

# Function to find the best match for each city
def normalize_city(city, choices, threshold=80):
    match = process.extractOne(city, choices)
    if match and match[1] >= threshold:
        return match[0]  # Return the matched city
    return city  # If no match is found or the match is below threshold, return the original

# Normalizing city names using fuzzy matching
df_pandas['normalized_city'] = df_pandas['city'].apply(lambda x: normalize_city(x, standardized_cities))

# Displaying the result
print(df_pandas[['city', 'normalized_city']].drop_duplicates())
```
